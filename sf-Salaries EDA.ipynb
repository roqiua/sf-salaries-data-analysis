Replace "**Not Provided**" with **NaN** in columns that should contain **numeric values**, then cast those columns to **float**.

```python
df["BasePay"] = df["BasePay"].replace("Not Provided", np.nan)
df["BasePay"] = pd.to_numeric(df["BasePay"])
```

```python
df["OvertimePay"] = df["OvertimePay"].replace("Not Provided", np.nan)
df["OvertimePay"] = pd.to_numeric(df["OvertimePay"])
```

```python
df["OtherPay"] = df["OtherPay"].replace("Not Provided", np.nan)
df["OtherPay"] = pd.to_numeric(df["OtherPay"])
```

```python
df["Benefits"] = df["Benefits"].replace("Not Provided", np.nan)
df["Benefits"] = pd.to_numeric(df["Benefits"])
```

```python
df.isna().sum()
```

Exclude rows where all three pay columns (**BasePay,OvertimePay,OtherPay**) are NaN, and rows where BasePay is NaN.

```python
mask = (
    ~(df["BasePay"].isna() &
      df["OvertimePay"].isna() &
      df["OtherPay"].isna())
    & df["BasePay"].notna()
)
df.loc[mask, ["BasePay", "OvertimePay", "OtherPay", "TotalPay"]]
```

```python
np.allclose(
    df.loc[mask, "TotalPay"] - (
        df.loc[mask, "OvertimePay"] + df.loc[mask, "OtherPay"]
    ),
    df.loc[mask, "BasePay"],
    equal_nan=True
)
```

Since these columns are related, check how many rows have missing values in all of them.

```python
len(df[["BasePay", "OvertimePay", "OtherPay", "TotalPay"]][(df["BasePay"].isna()) & (df["OvertimePay"].isna()) 
& (df["OtherPay"].isna())])
```

```python
df[(df["BasePay"].isna()) & (df["OvertimePay"].isna()) 
& (df["OtherPay"].isna())]
```

Remove rows where all related salary columns have missing values.

```python
df.dropna(
    subset=["BasePay", "OvertimePay", "OtherPay"],
    how="all",
    inplace=True
)
```

```python
df.isna().sum()
```

#### BasePay

Fill the missing values in **BasePay** using the following formula:

**BasePay = TotalPay − (OtherPay + OvertimePay)**

```python
df[["BasePay", "OvertimePay", "OtherPay", "Benefits" ,"TotalPay", "TotalPayBenefits"]][(df["BasePay"].isna())]
```

```python
df["BasePay"] = df["BasePay"].fillna(
    df["TotalPay"] - (df["OtherPay"] + df["OvertimePay"])
)
```

```python
df.isna().sum()
```

Verify that TotalPay equals the sum of BasePay, OvertimePay, and OtherPay for all rows.


```python
np.allclose(
    df["BasePay"] + df["OvertimePay"] + df["OtherPay"],
    df["TotalPay"],
    equal_nan=True
)
```

#### Benefits

```python
df[["Benefits" ,"TotalPay", "TotalPayBenefits"]][(df["Benefits"].isna())]
```

```python
mask_benefit = df["Benefits"].notna()
mask_benefit
```

Verify the relationship between the TotalPay, Benefits, and TotalPayBenefits columns:
**TotalPayBenefits = TotalPay + Benefits**

```python
np.allclose(
    df.loc[mask_benefit,"Benefits"] + df.loc[mask_benefit,"TotalPay"] ,
    df.loc[mask_benefit,"TotalPayBenefits"],
    equal_nan=True
)
```

```python
df["Benefits"] = df["Benefits"].fillna(
    df["TotalPayBenefits"] - df["TotalPay"]
)
```

```python
df.isna().sum()
```

#### Removing Invalid Salary Records

Records with **0** in all salary-related columns (`BasePay`, `OvertimePay`, `OtherPay`, `Benefits`, and `TotalPay`) were considered invalid for salary analysis. These records were identified, removed from the dataset, and the removal was verified.

```python
mask_zero = ((df["BasePay"] == 0) &
        (df["OvertimePay"] == 0) &
        (df["OtherPay"] == 0) &
        (df["TotalPay"] == 0) & 
        (df["Benefits"] == 0) )
len(df[mask_zero])
```

```python
df = df[~(mask_zero)]
```

```python
len(df[mask_zero])
```

#### Negative Salary Values

Negative values were identified in some salary-related columns. These records were retained because they may reflect payroll adjustments or accounting corrections rather than invalid data.

```python
mask_mins = ((df["BasePay"] < 0) |
        (df["OvertimePay"] < 0) |
        (df["OtherPay"] < 0) |
        (df["TotalPay"] < 0) | 
        (df["Benefits"] < 0) )
len(df[mask_mins])
```

```python
df[["EmployeeName","BasePay","OvertimePay" ,"OtherPay" ,"Benefits" ,"TotalPay", "TotalPayBenefits"]][mask_mins]
```

```python
df.isna().sum()
```

#### Removing Unnecessary Columns

The `Id`, `Notes`, `Status`, and `Agency` columns were removed because they are not relevant to the objectives of this analysis and do not provide useful information for salary exploration.

```python
df.drop(columns=["Notes","Id","Status","Agency"], inplace=True)
```
